```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
d1=pd.read_csv("/content/Rice Exporters 2019 - Trade Atlas Records - Jan.csv")
```

```python
d2 = pd.read_csv("/content/Rice Exporters 2020 - Jan.csv")
```

```python
d3=pd.read_csv("/content/Rice Exporters 2021 - January.csv")
```

```python
d4=pd.read_csv("/content/Rice Exporters 2022 - Jan.csv")
```

```python
df = [d1,d2,d3,d4]
df = reduce(lambda left,right: pd.merge(left,right,how='outer'),df)
```
